```python
import pandas as pd
import numpy as np
```

```python
file_path = "Raw/model_linkedin_data_content.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
for index, row in df.iterrows():
    df.at[index, 'Impressões (orgânicas)'] = np.random.randint(100, 1000)
    df.at[index, 'Impressões (patrocinadas)'] = np.random.randint(50, 500)
    df.at[index, 'Impressões (total)'] = df.at[index, 'Impressões (orgânicas)'] + df.at[index, 'Impressões (patrocinadas)']
    
    df.at[index, 'Impressões únicas (orgânicas)'] = np.random.randint(0, df.at[index, 'Impressões (total)'])
    
    df.at[index, 'Cliques (orgânicos)'] = np.random.randint(0, df.at[index, 'Impressões (orgânicas)'])
    df.at[index, 'Cliques (patrocinados)'] = np.random.randint(0, df.at[index, 'Impressões (patrocinadas)'])
    df.at[index, 'Cliques (total)'] = df.at[index, 'Cliques (orgânicos)'] + df.at[index, 'Cliques (patrocinados)']
    
    df.at[index, 'Reações (orgânicas)'] = np.random.randint(0, df.at[index, 'Comentários (total)'] + 1)
    df.at[index, 'Reações (patrocinadas)'] = np.random.randint(0, df.at[index, 'Comentários (total)'] + 1)
    df.at[index, 'Reações (total)'] = max(1, df.at[index, 'Reações (orgânicas)'] + df.at[index, 'Reações (patrocinadas)'])
    
    df.at[index, 'Comentários (orgânicos)'] = np.random.randint(0, df.at[index, 'Impressões (total)'])
    df.at[index, 'Comentários (patrocinados)'] = np.random.randint(0, df.at[index, 'Impressões (total)'])
    df.at[index, 'Comentários (total)'] = df.at[index, 'Comentários (orgânicos)'] + df.at[index, 'Comentários (patrocinados)']
    
    df.at[index, 'Compartilhamentos (orgânicos)'] = np.random.randint(0, df.at[index, 'Reações (total)'])
    df.at[index, 'Compartilhamentos (patrocinados)'] = np.random.randint(0, df.at[index, 'Reações (total)'])
    df.at[index, 'Compartilhamentos (total)'] = df.at[index, 'Compartilhamentos (orgânicos)'] + df.at[index, 'Compartilhamentos (patrocinados)']
    
    df.at[index, 'Taxa de engajamento (orgânico)'] = np.random.uniform(0, 0.1)
    df.at[index, 'Taxa de engajamento (patrocinado)'] = np.random.uniform(0, 0.1)
    df.at[index, 'Taxa de engajamento (total)'] = df.at[index, 'Taxa de engajamento (orgânico)'] + df.at[index, 'Taxa de engajamento (patrocinado)']

df
```

```python
df.to_csv('Raw/model_linkedin_data_content_populated.csv')
```
